```python
#Proyecto 6
#Hacemos una calculadora
#Autor Misael
#Version 1.0
#Fecha

def menu():
	print("Bienvenido a la calculadora 5000")
	print("1.-Sumar")
	print("2.-Restar")
	print("3.-Multiplicar")
	print("4.-Dividir")
	print("5.-Salir")

def sumar(valorA, valorB):
	resultado = valorA+valorB
	print(f'El resultado es {resultado}')

def restar(valor1, valor2):
	resultado = valorA-valorB
	print(f'El resultado es {resultado}')

def multiplicar(valor1, valor2):
	resultado = valorA*valorB
	print(f'El resultado es {resultado}')

def dividir(valorA, valorB):
	resultado = valorA/valorB
	print(f'El resultado es {resultado}')

#------------------------------------------                                     #Calculadora con operaciones aritmeticas simples
menu()
repetirprograma=True
while (repetirprograma==True):

	opcion=int(input())
	if(opcion!=5):
		print("Ingrese el primer valor")
		valorA=int(input())
		print("Ingrese el segundo valor")
		valorB=int(input())

	if (opcion==1):
		sumar(valorA,valorB)
	elif(opcion==2):
		restar(valorA,valorB)
	elif(opcion==3):
		multiplicar(valorA,valorB)
	elif(opcion==4):
		dividir(valorA,valorB)
	elif(opcion==5):
		repetirprograma=False
```
